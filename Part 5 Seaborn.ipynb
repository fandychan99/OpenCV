### Seaborn dan Pandas DataFrame

```python
df = pd.read_csv('data/cereal.csv', index_col='name')
df.head()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.isna(), cmap='Greens');
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.isna(), cmap='Greens');
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(numeric_only=True), cmap='seismic', vmin=-1, vmax=1, annot=True, square=True);
```
